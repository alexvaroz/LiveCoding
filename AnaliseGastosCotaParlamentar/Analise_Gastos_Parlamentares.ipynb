Perguntas interessantes:

* Qual parlamentar gastou mais?
* Qual partido gastou mais?
* Qual parlamentar de Brasília gastou mais?
* Qual a média de gastos por deputado?
* Qual foi o maior valor de nota fiscal?
* Qual o menor valor?
* Qual parlamentar gastou menos?
* Qual a média em cada categoria de despesa?

```python
# Qual parlamentar gastou mais??
agrupados = df.groupby(['txNomeParlamentar', 'sgPartido', 'sgUF'])['vlrDocumento'].sum().sort_values(ascending = False).head(10)
```

```python
# Qual parlamentar de Brasília gastou mais??
df[df['sgUF'] == 'DF'].groupby('txNomeParlamentar')['vlrDocumento'].sum().sort_values(ascending = False)
```

```python
# Como foi o gasto da Bia Kicis?
df[df['txNomeParlamentar'] == 'Bia Kicis'].groupby('txtDescricao')['vlrDocumento'].sum()
```

```python
# Qual o maior valor de nota fiscal?
df[['txNomeParlamentar', 'sgPartido', 'sgUF', 
    'txtDescricao', 'vlrDocumento']].sort_values('vlrDocumento', ascending = False)
```
